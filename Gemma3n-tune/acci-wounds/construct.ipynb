````python
# Prepare to read already processed images from both CSV and JSONL files
processed_images = set()

# Define the path for the old CSV file
OLD_CSV_FILE = os.path.join(drive_base_path, "datatillnow.csv")
OUTPUT_JSONL = os.path.join(drive_base_path, "dataset.jsonl")
# Read processed images from the old CSV file if it exists
if os.path.exists(OLD_CSV_FILE):
    print(f"Reading processed images from old CSV file: {OLD_CSV_FILE}...")
    try:
        with open(OLD_CSV_FILE, "r", newline='', encoding='utf-8') as f:
            reader = csv.reader(f)
            next(reader, None)  # skip header
            for row in reader:
                if row and len(row) > 0:
                    processed_images.add(row[0])
    except FileNotFoundError:
        print(f"Old CSV file not found: {OLD_CSV_FILE}")
    except Exception as e:
        print(f"Error reading old CSV file {OLD_CSV_FILE}: {e}")


# Read processed images from the new JSONL file if it exists
if os.path.exists(OUTPUT_JSONL):
    print(f"Reading processed images from JSONL file: {OUTPUT_JSONL}...")
    try:
        with open(OUTPUT_JSONL, "r", encoding='utf-8') as f:
            for line in f:
                try:
                    entry = json.loads(line)
                    if "filename" in entry:
                        processed_images.add(entry["filename"])
                except json.JSONDecodeError as e:
                    # Print a more informative message for JSON errors
                    print(f"Skipping invalid JSON line in {OUTPUT_JSONL}: {line.strip()} - Error: {e}")
                except Exception as e:
                    print(f"Error processing line in {OUTPUT_JSONL} ({line.strip()}): {e}")
    except FileNotFoundError:
        print(f"JSONL file not found: {OUTPUT_JSONL}")
    except Exception as e:
        print(f"Error reading JSONL file {OUTPUT_JSONL}: {e}")

print(f"Found a total of {len(processed_images)} already processed images from CSV and JSONL.")

# Initialize an empty list to store results for the current run (optional, could write directly to file)
# results_list = [] # We will write directly to JSONL instead of building a list in memory
````

## Adjust data extraction

### Subtask:
Modify the `process_image` function to extract the necessary data fields (e.g., image path, user query, assistant response) in the format required for fine-tuning.


**Reasoning**:
Modify the `process_image` function to return a dictionary with the required fields for fine-tuning data, including the image filename and a list of messages in a chat format.



````python
def extract_number(filename):
    match = re.search(r'(\d+)', filename)
    return int(match.group(1)) if match else float('inf')

def process_image(img_name):
    if img_name in processed_images:
        print(f"Skipping {img_name}, already processed.")
        return None
    image_path = os.path.join(image_folder, img_name)
    try:
        with open(image_path, "rb") as img_file:
            image_bytes = img_file.read()

        response = ollama.chat(
            model='llava',
            messages=[
                {
                    "role": "user",
                    "content": prompt_template,
                    "images": [image_bytes]
                }
            ]
        )

        json_output = response['message']['content'].strip()
        if not json_output:
            print(f"Empty response for {img_name}, skipping.")
            return {"filename": img_name, "error": "Empty response"}

        # Remove code block markers if present
        if json_output.startswith("```"):
            json_output = json_output.strip().lstrip("`")
            lines = json_output.splitlines()
            if lines and lines[0].strip().lower() == "json":
                lines = lines[1:]
            if lines and lines[-1].strip() == "```":
                lines = lines[:-1]
            json_output = "\n".join(lines).strip()

        # Parse JSON output
        try:
            parsed = json.loads(json_output)

            # Format messages for fine-tuning
            messages = [
                {"role": "user", "content": parsed.get('user-query', 'N/A')}, # Added N/A fallback
                {"role": "assistant", "content": parsed.get('assistant-response', 'N/A')} # Added N/A fallback
            ]

            print(f"\n \u2728 Processed {img_name} successfully.")
            return {
                "filename": img_name,
                "messages": messages,
                "injury-severity": parsed.get('injury-severity', 'Unknown') # Include severity if needed for finetuning, added Unknown fallback
            }

        except json.JSONDecodeError as json_e:
            print(f"JSON Decode Error processing {img_name}: {json_e}")
            print(f"Problematic JSON output: {json_output}") # Print the problematic output
            return {"filename": img_name, "error": f"JSON Decode Error: {json_e}", "problematic_output": json_output} # Include problematic output in error dict
        except Exception as other_e:
            print(f"Error processing {img_name} after AI response: {other_e}")
            print(f"Problematic JSON output: {json_output}") # Print the problematic output
            return {"filename": img_name, "error": str(other_e), "problematic_output": json_output} # Include problematic output in error dict

    except Exception as e:
        print(f"Error processing {img_name}: {e}")
        return {"filename": img_name, "error": str(e)}
````

## Modify data saving

### Subtask:
Change the data saving mechanism from writing to a CSV file to writing to a file in the required fine-tuning format (e.g., JSONL).


**Reasoning**:
Update the main processing loop to save the results to a JSONL file instead of a CSV file, handling errors and skipping already processed images.



````python
# Define the output JSONL file path
jsonl_file = os.path.join(drive_base_path, "dataset.jsonl")

# Create the images folder if it doesn't exist (already done in setup, but good to be safe)
if not os.path.exists(image_folder):
    os.makedirs(image_folder)
    print(f"Created folder: {image_folder}")

# Get a list of all files in the image_folder.
all_files = os.listdir(image_folder)

# Filter for image files and those not already processed
image_list = [f for f in all_files if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')) and f not in processed_images]

# Sort images if necessary (optional, based on extract_number function)
image_list.sort(key=extract_number)

# Remove the logic to start from img1000.jpg
# try:
#     start_index = image_list.index('img1000.jpg')
#     image_list = image_list[start_index:]
#     print(f"Starting processing from img1000.jpg. {len(image_list)} images remaining.")
# except ValueError:
#     print("img1000.jpg not found in the list of unprocessed images. Processing all remaining images.")


err = []

print(f"Found {len(image_list)} new images to process.")

# --- Image Processing Loop with ThreadPoolExecutor ---
# Use all available CPU cores for parallel processing
max_workers = os.cpu_count() or 4 # Use a default of 4 if CPU count is not available

# Open the JSONL file in append mode at the beginning
with open(jsonl_file, "a", encoding='utf-8') as f:
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = {executor.submit(process_image, img_name): img_name for img_name in image_list}

        for future in tqdm(as_completed(futures), total=len(futures), desc="Processing Images"):
            img_name = futures[future]
            try:
                result = future.result()

                if result is None:
                    # Skipped because it was already processed (handled by process_image)
                    continue

                # Check if the result contains an error
                if "error" in result:
                    err.append((result["filename"], result["error"]))
                else:
                    # Write the successful result to the JSONL file
                    json.dump(result, f)
                    f.write('\n') # Add a newline for each JSON object
                    # Add to processed_images set to avoid reprocessing in case of interruption
                    # Although process_image already checks this, adding here provides an extra layer
                    processed_images.add(result["filename"])

            except Exception as exc:
                err.append((img_name, f'Generated an exception: {exc}'))


print("Finished processing all files in the directory.")
if err:
    print(f"Some errors occurred during processing: {err}")
    print(f"Details of errors: {err}")

print(f"Results saved in {jsonl_file}")
````

## Add data validation

### Subtask:
Include steps to validate the generated data to ensure it meets the requirements for fine-tuning.


**Reasoning**:
Implement the data validation steps as instructed, reading the generated JSONL file, iterating through each entry, and checking for the required keys and message structure.



````python
# Validate the generated dataset.jsonl file
print("\n--- Validating dataset.jsonl ---")

valid_entries_count = 0
invalid_entries = []

# Define the expected keys for each entry
expected_entry_keys = ["filename", "messages"]
# Define the expected keys for each message within the messages list
expected_message_keys = ["role", "content"]

try:
    with open(jsonl_file, "r", encoding='utf-8') as f:
        for line_num, line in enumerate(f):
            try:
                entry = json.loads(line)
                is_valid = True
                validation_errors = []

                # Check for required top-level keys
                for key in expected_entry_keys:
                    if key not in entry:
                        is_valid = False
                        validation_errors.append(f"Missing key: '{key}'")

                if is_valid and not isinstance(entry["messages"], list):
                    is_valid = False
                    validation_errors.append("'messages' is not a list")

                if is_valid:
                    messages = entry["messages"]
                    if not messages:
                        is_valid = False
                        validation_errors.append("'messages' list is empty")
                    else:
                        has_user_message = False
                        has_assistant_message = False
                        for msg_num, message in enumerate(messages):
                            if not isinstance(message, dict):
                                is_valid = False
                                validation_errors.append(f"Message at index {msg_num} is not a dictionary")
                                continue

                            for key in expected_message_keys:
                                if key not in message:
                                    is_valid = False
                                    validation_errors.append(f"Message at index {msg_num} missing key: '{key}'")

                            if is_valid:
                                if message.get("role") == "user":
                                    has_user_message = True
                                elif message.get("role") == "assistant":
                                    has_assistant_message = True

                        # Optional: Check if at least one user and one assistant message exist
                        # if not has_user_message:
                        #     is_valid = False
                        #     validation_errors.append("Missing user message in 'messages'")
                        # if not has_assistant_message:
                        #     is_valid = False
                        #     validation_errors.append("Missing assistant message in 'messages'")


                if is_valid:
                    valid_entries_count += 1
                else:
                    invalid_entries.append({"line_num": line_num + 1, "filename": entry.get("filename", "N/A"), "errors": validation_errors})

            except json.JSONDecodeError as e:
                invalid_entries.append({"line_num": line_num + 1, "filename": "N/A (JSON Decode Error)", "errors": [f"JSON Decode Error: {e}"]})
            except Exception as e:
                invalid_entries.append({"line_num": line_num + 1, "filename": "N/A", "errors": [f"Unexpected error: {e}"]})

except FileNotFoundError:
    print(f"Error: The file {jsonl_file} was not found.")
except Exception as e:
    print(f"Error reading file {jsonl_file}: {e}")


print(f"\n--- Validation Summary ---")
print(f"Total entries processed: {valid_entries_count + len(invalid_entries)}")
print(f"Valid entries: {valid_entries_count}")
print(f"Invalid entries: {len(invalid_entries)}")

if invalid_entries:
    print("\nDetails of Invalid Entries:")
    for entry_info in invalid_entries:
        print(f"  Line {entry_info['line_num']} (Filename: {entry_info['filename']}):")
        for error in entry_info['errors']:
            print(f"    - {error}")
else:
    print("\nAll entries are valid according to the defined checks.")
````
